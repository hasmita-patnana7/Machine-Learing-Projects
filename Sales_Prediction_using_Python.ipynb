Impoting Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Dataset

```python
df=pd.read_csv("Advertising.csv")
df.head()
```

```python
df.drop('Unnamed: 0',axis=1, inplace= True)
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x=df['TV'],y=df['Sales'])
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x=df['Radio'],y=df['Sales'])
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x=df['Newspaper'],y=df['Sales'])
```

```python
X = df.iloc[: ,:-1]
Y = df.iloc[: ,-1]
X.head()
Y.head()
```

Splitting the dataset into training and test sets

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train ,Y_test = train_test_split(X , Y , test_size=0.3 , random_state=101)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print("X_train: ")
print(X_train.head())
```

```python
print("X_test: ")
print(X_test.head())
```

```python
print("Y_train: ")
print(Y_train.head())
```

```python
print("Y_test: ")
print(Y_test.head())
```

Regressor

```python
from xgboost import XGBRegressor
model=XGBRegressor()
```

Training the Model on the dataset

```python
model.fit(X_train  , Y_train)
```

```python
pred=model.predict(X_test)    # mean
```

Evaluating the Model

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error ,r2_score
m = mean_absolute_error(Y_test , pred)
m = mean_squared_error(Y_test , pred)
r2 = r2_score(Y_test , pred)
```

```python
print("Mean absolute error: {}".format(round(m,3)))
print("Mean squared error: {}".format(round(m,3)))
```

```python
print("r2 score: {}".format(round(r2,3)))
```

Testing the model

```python
tv = float(input("Cost of TV : "))
radio = float(input("Cost of Radio : "))
newspaper = float(input("Cost of newspaper : "))
```

```python
arr = [tv , radio , newspaper]
arr = np.array(arr)
arr = arr.reshape(1 , -1)
```

```python
result = model.predict(arr)
```

```python
print("Sales : {}".format(result))
```
